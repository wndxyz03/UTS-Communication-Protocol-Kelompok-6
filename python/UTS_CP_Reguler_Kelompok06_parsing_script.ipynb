```python
import json
import pandas as pd

postman_data = {
    "info": {
        "name": "New Collection",
        "schema": "https://schema.getpostman.com/json/collection/v2.1.0/collection.json",
    },
    "item": [
        {
            "name": "POST 1 Users",
            "request": {
                "method": "POST",
                "body": {
                    "mode": "raw",
                    "raw": '{\r\n  "username": "azka",\r\n  "email": "azka@gmail.com",\r\n  "role": "1",\r\n  "status": "active",\r\n  "department": "Kelompok 6"\r\n}',
                },
            },
        },
        {
            "name": "POST 2 Profiles",
            "request": {
                "method": "POST",
                "body": {
                    "mode": "raw",
                    "raw": '{\r\n  "userId": 1,\r\n  "fullName": "Kanaya",\r\n  "program": "Sains Data",\r\n  "semester": 2,\r\n  "skills": [\r\n    "postman",\r\n    "json",\r\n    "wireshark"\r\n  ],\r\n  "city": "Jakarta"\r\n}',
                },
            },
        },
    ],
}

for item in postman_data["item"]:
    # 1. Ambil nama item (untuk nama file & print) dan load raw string ke dict
    name = item["name"].split()[-1].lower()  # Menghasilkan 'users' atau 'profiles'
    data_dict = json.loads(item["request"]["body"]["raw"])

    # 2. Gabungkan list skills jika ada
    if "skills" in data_dict:
        data_dict["skills"] = ", ".join(data_dict["skills"])

    # 3. Buat DataFrame, Tampilkan, dan Export ke CSV
    df = pd.DataFrame([data_dict])

    print(f"\n=== TABEL DATA {name.upper()} ===")
    print(df)

    df.to_csv(f"output_{name}.csv", index=False)

print("\nFile 'output_users.csv' dan 'output_profiles.csv' berhasil disimpan!")
```
